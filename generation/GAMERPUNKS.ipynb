```python
from PIL import Image, ImageOps
from IPython.display import display
import random
import json
import math

setHair = {
    "Blue Normal": 20,
    "Green Clown": 20,
    "Green Vegeta": 20,
    "Light Blue Flat": 10,
    "Rainhawkbow": 10,
    "Red Braids": 10,
    "Yellow Trump": 10,
}
attrHair = list(setHair.keys())
weightHair = list(setHair.values())
if (int(sum(weightHair)) != 100): raise ValueError('weightHair: ', sum(weightHair))

setEar = {
    "Normal": 33,
    "Red Headset": 67,
}
attrEar = list(setEar.keys())
weightEar = list(setEar.values())
if (int(sum(weightEar)) != 100): raise ValueError('weightEar: ', sum(weightEar))

setBrow = {
    "Normal": 100,
}
attrBrow = list(setBrow.keys())
weightBrow = list(setBrow.values())
if (int(sum(weightBrow)) != 100): raise ValueError('weightBrow: ', sum(weightBrow))

setEye = {
    "Normal": 33,
    "VR Set": 67,
}
attrEye = list(setEye.keys())
weightEye = list(setEye.values())
if (int(sum(weightEye)) != 100): raise ValueError('weightEye: ', sum(weightEye))

setNose = {
    "Normal": 100,
}
attrNose = list(setNose.keys())
weightNose = list(setNose.values())
if (int(sum(weightNose)) != 100): raise ValueError('weightNose: ', sum(weightNose))


setMouth = {
    "Normal": 70,
    "Smoking Pipe": 30,
}
attrMouth = list(setMouth.keys())
weightMouth = list(setMouth.values())
if (int(sum(weightMouth)) != 100): raise ValueError('weightMouth: ', sum(weightMouth))



setShirt = {
    "Blue Jacket": 100
}
attrShirt = list(setShirt.keys())
weightShirt = list(setShirt.values())
if (int(sum(weightShirt)) != 100): raise ValueError('weightShirt: ', sum(weightShirt))

setBase = {
    'Default':100,
}
attrBase = list(setBase.keys())
weightBase = list(setBase.values())
if (int(sum(weightBase)) != 100): raise ValueError('weightBase: ', sum(weightBase))

setChair = {
    'Green Ergotune':100,
}
attrChair = list(setChair.keys())
weightChair = list(setChair.values())
if (int(sum(weightChair)) != 100): raise ValueError('weightChair: ', sum(weightChair))

setDesktop = {
    '4pp13':100,
}
attrDesktop = list(setDesktop.keys())
weightDesktop = list(setDesktop.values())
if (int(sum(weightDesktop)) != 100): raise ValueError('weightDesktop: ', sum(weightDesktop))

setSpecial = {
    "Karambit": 100,
}
attrSpecial = list(setSpecial.keys())
weightSpecial = list(setSpecial.values())
if (int(sum(weightSpecial)) != 100): raise ValueError('weightSpecial: ', sum(weightSpecial))


attributeMap = {
    "Hair": attrHair,
    "Ear": attrEar,
    "Brow": attrBrow,
    "Eye": attrEye,
    "Nose": attrNose,
    "Mouth": attrMouth,
    "Shirt": attrShirt,
    "Base": attrBase,
    "Chair": attrChair,
    "Desktop": attrDesktop,
    "Special": attrSpecial,
}
weightMap = {
    "Hair": weightHair,
    "Ear": weightEar,
    "Brow": weightBrow,
    "Eye": weightEye,
    "Nose": weightNose,
    "Mouth": weightMouth,
    "Shirt": weightShirt,
    "Base": weightBase,
    "Chair": weightChair,
    "Desktop": weightDesktop,
    "Special": weightSpecial,
}

filesHair = {
    "Blue Normal":"Blue Normal.png",
    "Green Clown":"Green Clown.png",
    "Green Vegeta":"Green Vegeta.png",
    "Light Blue Flat":"Light Blue Flat.png",
    "Rainhawkbow":"Rainhawkbow.png",
    "Red Braids":"Red Braids.png",
    "Yellow Trump":"Yellow Trump.png",
}
filesEar = {
    "Normal":"Normal.png",
    "Red Headset": "Red Headset.png",
}
filesEye = {
    "Normal": "Normal.png",
    "VR Set": "VR Set.png",
}
filesBrow = {
    "Normal": "Normal.png",
}
filesNose = {
    "Normal": "Normal.png",
}
filesShirt = {
    "Blue Jacket": "Blue Jacket.png",
}
filesMouth = {
    "Normal": "Normal.png",
    "Smoking Pipe": "Smoking Pipe.png",
}
filesBase = {
    'Default':'Default.png',
}
filesChair = {
    "Green Ergotune": "Green Ergotune.png",
}
filesDesktop = {
    "4pp13": "4pp13.png",
}
filesSpecial = {
    "Karambit": "Karambit.png",
}

```

```python
## TRAIT GENERATION

TOTAL_GAMERPUNKS = 369

traits = []


attributeCount = [0,1,2,3,4,]
attributeCountWeight = [1,5,29,65,1]


attributes = ["Hair","Brow","Shirt","Special"]
attributeSet = set(attributes)
attributeSetWeight = {
    "Hair": 133,
    "Brow": 199,
    "Shirt": 166,
    "Special": 1,
}
asw = attributeSetWeight

def createCombo():
    trait = {}
    traitAttributed = []

    trait["AttributeCount"] = random.choices(attributeCount, attributeCountWeight)[0]

    trait["Chair"] = random.choices(attrChair, weightChair)[0]
    trait["Desktop"] = random.choices(attrDesktop, weightDesktop)[0]
    trait["Base"] = random.choices(attrBase, weightBase)[0]
    trait["Ear"] = random.choices(attrEar, weightEar)[0]
    trait["Eye"] = random.choices(attrEye, weightEye)[0]
    trait["Nose"] = random.choices(attrNose, weightNose)[0]
    trait["Mouth"] = random.choices(attrMouth, weightMouth)[0]
    
    # set traits to add
    for i in range(trait["AttributeCount"]):
        # remove from set of attributes, the already added attributes.
        possibleAttributes = list(attributeSet.symmetric_difference(traitAttributed))
        # update weight of new set
        attributesWeight = []
        totalAttributeWeight = 0
        for j in possibleAttributes:
            totalAttributeWeight += asw[j]
        for j in possibleAttributes:
            attributesWeight.append(asw[j] / totalAttributeWeight * 100)
        traitAttributed.append(random.choices(possibleAttributes, attributesWeight)[0])
        
    for i in traitAttributed:
        trait[i] = random.choices(attributeMap[i], weightMap[i])[0]

    # trait["Neck"] = random.choices(attrNeck, weightNeck)[0]
    # trait["Mouth"] = random.choices(attrMouth, weightMouth)[0]
    # trait["Eye"] = random.choices(attrEye, weightEye)[0]
    # trait["Ear"] = random.choices(attrEar, weightEar)[0]
    # trait["Mask"] = random.choices(attrMask, weightMask)[0]
    # trait["Ant"] = random.choices(attrAnt, weightAnt)[0]
    # trait["Scar"] = random.choices(attrScar, weightScar)[0]
    

    # if trait in traits:
    #     return createCombo()
    # else:
    #   return trait
    return trait
    
for i in range(TOTAL_GAMERPUNKS):
    
    newtraitcombo = createCombo()
    
    traits.append(newtraitcombo)
    
```

```python
#picture collage
import math

picsize = 100
xwidth = 1920
ywidth = math.floor(TOTAL_GAMERPUNKS / (xwidth / picsize)) * picsize

collage = Image.new("RGBA", (xwidth,ywidth))

xcounter = 0
ycounter = 0

SQUARE_RATIO = 500
for item in traits:
    images = []

    attributeCount = item["AttributeCount"]
 
    images.append(Image.new("RGBA", (SQUARE_RATIO,SQUARE_RATIO),(random.randint(0,20),random.randint(0,20),random.randint(0,20),255))) # (0,0,0,255))) #
    if("Chair" in item): images.append(Image.open(f'./input/chair/{filesChair[item["Chair"]]}').convert('RGBA'))
    if("Desktop" in item): images.append(Image.open(f'./input/desktop/{filesDesktop[item["Desktop"]]}').convert('RGBA'))
    if("Base" in item): images.append(Image.open(f'./input/base/{filesBase[item["Base"]]}').convert('RGBA'))
    if("Brow" in item): images.append(Image.open(f'./input/brow/{filesBrow[item["Brow"]]}').convert('RGBA'))
    if("Nose" in item): images.append(Image.open(f'./input/nose/{filesNose[item["Nose"]]}').convert('RGBA'))
    if("Eye" in item): images.append(Image.open(f'./input/eye/{filesEye[item["Eye"]]}').convert('RGBA'))
    if("Mouth" in item): images.append(Image.open(f'./input/mouth/{filesMouth[item["Mouth"]]}').convert('RGBA'))
    if("Ear" in item): images.append(Image.open(f'./input/ear/{filesEar[item["Ear"]]}').convert('RGBA'))
    if("Special" in item): images.append(Image.open(f'./input/special/{filesSpecial[item["Special"]]}').convert('RGBA'))
    if("Hair" in item): images.append(Image.open(f'./input/hair/{filesHair[item["Hair"]]}').convert('RGBA'))
    if("Shirt" in item): images.append(Image.open(f'./input/shirt/{filesShirt[item["Shirt"]]}').convert('RGBA'))

    composite = images.pop(0)
    for image in images:
        composite = Image.alpha_composite(composite, image)

    #Convert to RGB
    # rgb_im = com8.convert('RGB')
    collage.paste(composite.resize((picsize,picsize), Image.NEAREST), (xcounter, ycounter))
    xcounter += picsize
    if(xcounter >= xwidth):
        xcounter = 0
        ycounter += picsize
    
    
    composite = composite.resize((SQUARE_RATIO,SQUARE_RATIO), Image.NEAREST)

    file_name = str(item["tokenId"]) + ".png"
    composite.save("./output/" + file_name)
    
    if(int(item["tokenId"]) < 60):
        file_name = str(item["tokenId"]) + ".jpg"
        bg = Image.new("RGB", composite.size, (255,255,255))
        bg.paste(composite,composite)
        bg.save("./outputgif/" + file_name)

collageFileName = "collage.png"
collage.save("./" + collageFileName)
```
